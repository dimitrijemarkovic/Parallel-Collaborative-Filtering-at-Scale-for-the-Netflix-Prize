```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error, accuracy_score
import joblib
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
train_file_path = r'C:\Users\Dida\Downloads\train_features.csv'
train_data = pd.read_csv(train_file_path)


X_train = train_data.drop(columns=['target'])  
y_train = train_data['target'] 
```

```python
ridge = Ridge()
parameters = {'alpha': [0.1, 1.0, 10.0, 100.0, 200.0]}
```

```python
grid_search = GridSearchCV(ridge, parameters, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(X_train, y_train)
```

```python
y_train_pred = grid_search.predict(X_train)


train_rmse = mean_squared_error(y_train, y_train_pred, squared=False)
print(f'Training RMSE: {train_rmse}')
```

```python
model_path = 'ridge_model.pkl'
joblib.dump(grid_search.best_estimator_, model_path)


loaded_model = joblib.load(model_path)


test_file_path = r'C:\Users\Dida\Downloads\val_features.csv' 
test_data = pd.read_csv(test_file_path)


X_test = test_data.drop(columns=['target'])  
y_test = test_data['target'] 


y_pred = loaded_model.predict(X_test)


rmse = mean_squared_error(y_test, y_pred, squared=False)
print(f'Test RMSE: {rmse}')
```

```python
coefficients = loaded_model.coef_
features = X_train.columns

coef_df = pd.DataFrame({
    'Feature': features,
    'Coefficient': coefficients
})

coef_df['AbsCoefficient'] = np.abs(coef_df['Coefficient'])
coef_df = coef_df.sort_values(by='AbsCoefficient', ascending=False)

print(coef_df)
```

```python
plt.figure(figsize=(10, 6))
plt.barh(coef_df['Feature'], coef_df['Coefficient'])
plt.xlabel('Coefficient')
plt.title('Feature Importance')
plt.gca().invert_yaxis()
plt.show()
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
ridge = Ridge()
grid_search = GridSearchCV(ridge, parameters, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(X_train_scaled, y_train)
```

```python
coefficients = grid_search.best_estimator_.coef_
coef_df = pd.DataFrame({
    'Feature': X_train.columns,
    'Coefficient': coefficients,
    'AbsCoefficient': np.abs(coefficients)
}).sort_values(by='AbsCoefficient', ascending=False)

print(coef_df)
```

```python
plt.figure(figsize=(10, 6))
plt.barh(coef_df['Feature'], coef_df['Coefficient'])
plt.xlabel('Coefficient')
plt.title('Feature Importance')
plt.gca().invert_yaxis()
plt.show()
```

```python
y_pred = grid_search.predict(X_test_scaled)
rmse = mean_squared_error(y_test, y_pred, squared=False)
print(f'Test RMSE: {rmse}')
```

```python
# Izbacivanje atributa sa najmanjim koeficijentima
X_train_reduced = X_train.drop(columns=['movie_mean_rating', 'user_rating_variability'])
X_test_reduced = X_test.drop(columns=['movie_mean_rating', 'user_rating_variability'])

X_train_scaled_reduced = scaler.fit_transform(X_train_reduced)
X_test_scaled_reduced = scaler.transform(X_test_reduced)
```

```python
ridge = Ridge()
grid_search = GridSearchCV(ridge, parameters, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(X_train_scaled_reduced, y_train)
```

```python
coefficients = grid_search.best_estimator_.coef_
coef_df_reduced = pd.DataFrame({
    'Feature': X_train_reduced.columns,
    'Coefficient': coefficients,
    'AbsCoefficient': np.abs(coefficients)
}).sort_values(by='AbsCoefficient', ascending=False)

print(coef_df_reduced)
```

```python
plt.figure(figsize=(10, 6))
plt.barh(coef_df_reduced['Feature'], coef_df_reduced['Coefficient'])
plt.xlabel('Coefficient')
plt.title('Feature Importance')
plt.gca().invert_yaxis()
plt.show()
```

```python
y_pred_reduced = grid_search.predict(X_test_scaled_reduced)
rmse_reduced = mean_squared_error(y_test, y_pred_reduced, squared=False)
print(f'Test RMSE after removing features: {rmse_reduced}')
```

### Analiza važnosti atributa

Nakon treniranja Ridge regresije, analizirali smo koeficijente atributa kako bismo utvrdili njihov značaj za model. Najvažniji atributi su:

- **knn_pred**: Ovaj atribut ima najveći koeficijent, što ga čini najuticajnijim faktorom u modelu.
- **user_mean_rating**: Ovaj atribut ima negativan uticaj, što znači da korisnici koji generalno daju više ocene imaju tendenciju da smanjuju predikciju modela.
- **svd_pred**: Takođe doprinosi pozitivno konačnoj predikciji, ali manje u odnosu na KNN predikciju.

Testirali smo standardizaciju atributa i uklanjanje manje bitnih atributa, ali ti postupci nisu doveli do značajnog poboljšanja RMSE (0.7249). Ovaj rezultat ukazuje na to da su predikcije modela KNN i SVD, kao i prosečna ocena korisnika, ključni faktori za tačnost modela.
